```python
import numpy as np
import matplotlib.pyplot as plt
import uncertainties as uc
```

```python
F = uc.ufloat(2.3,.05)
L = uc.ufloat(7.17e-3, .05e-3)
w = uc.ufloat(.63e-3,.005e-3)
t =uc.ufloat (.16e-3, .005e-3)
lamb = 650
E = 2.08e12
```

```python
dy_theory_eq = lambda x: (12*F/(E*w*t**3)) * (L**3/6 - (L-x)**3/6 - L**2*x/2)

#example calc: for all positions of fringes, x 
# x approximated from image
x = np.array([5,9,12.5,15,17,19,20.5,21.5])*1e-3

dy_t = dy_theory_eq(x)
plt.plot(x, [i.n for i in dy_t])
plt.plot(x, [i.n for i in dy_t], '.')
plt.xlabel('x (m)')
plt.ylabel('delta y(x)')
plt.title('Cantilever displacement for very approximate x')
plt.savefig('theory')
```
